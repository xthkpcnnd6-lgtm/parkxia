<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">SVM 기본 분류</span>

```python
from sklearn import datasets
from sklearn.svm import SVC

# 데이터
X, y = datasets.load_iris(return_X_y=True)

# 모델 생성
model = SVC(kernel='linear')

# 학습
model.fit(X, y)

# 예측
print(model.predict(X[:5]))
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">결정 경계(Decision Boundary) 개념</span>

```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

X, y = datasets.make_blobs(n_samples=100, centers=2, random_state=42)

model = SVC(kernel='linear')
model.fit(X, y)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Decision Boundary 개념")
plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">커널 트릭(Kernel Trick)</span>

```python
from sklearn.datasets import make_circles
from sklearn.svm import SVC

X, y = make_circles(noise=0.1, factor=0.5)

model = SVC(kernel='rbf')  # 비선형
model.fit(X, y)

print("커널 적용 완료")
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">마진(Margin) 개념</span>

```python
from sklearn.svm import SVC

model = SVC(kernel='linear', C=1)
model.fit(X, y)

print("마진 조절 C값:", model.C)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">C 값 (규제 강도)</span>

```python
model1 = SVC(C=0.1)   # 넓은 마진 (단순)
model2 = SVC(C=100)   # 좁은 마진 (정확도 ↑)

model1.fit(X, y)
model2.fit(X, y)

print("C 비교 완료")
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">서포트 벡터 확인</span>

```python
model = SVC(kernel='linear')
model.fit(X, y)

print("서포트 벡터 개수:", len(model.support_))
print("서포트 벡터 인덱스:", model.support_)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">SVM 회귀 (SVR)</span>

```python
from sklearn.svm import SVR
import numpy as np

X = np.arange(10).reshape(-1, 1)
y = X.ravel()

model = SVR()
model.fit(X, y)

print(model.predict([[5]]))
```
