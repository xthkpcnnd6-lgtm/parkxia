<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">선 그래프(line plot)</span>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y)
plt.title("Line Plot")
plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">막대 그래프(bar plot)</span>

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C']
values = [10, 30, 20]

plt.bar(labels, values)
plt.title("Bar Plot")
plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">히스토그램(histogram)</span>

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.hist(data)
plt.title("Histogram")
plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">산점도(scatter plot)</span>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">박스 플랏(box plot)</span>

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">히트맵(Heatmap - Seaborn)</span>

```python
import seaborn as sns
import numpy as np

data = np.random.rand(5, 5)

sns.heatmap(data, annot=True)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">여러 그래프 한 번에 (Subplot)</span>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))

plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">그래프 꾸미기 (핵심 옵션)</span>

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [3, 1, 4]

plt.plot(x, y, label="data")
plt.title("Title")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend()
plt.grid()

plt.show()
```
