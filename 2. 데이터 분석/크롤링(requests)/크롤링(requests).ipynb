<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">기본 요청 보내기(GET)</span>

```python
import requests

url = "https://example.com"
res = requests.get(url)

print(res.status_code)  # 200이면 성공
print(res.text[:200])   # HTML 일부 출력
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">상태 코드 확인하기</span>

```python
if res.status_code == 200:
    print("요청 성공")
else:
    print("요청 실패:", res.status_code)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">헤더 설정(차단 방지)</span>

```python
headers = {
    "User-Agent": "Mozilla/5.0"
}

res = requests.get(url, headers=headers)
print(res.status_code)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">쿼리 파라미터 보내기</span>

```python
params = {"q": "python"}

res = requests.get("https://www.google.com/search", params=params)
print(res.url)  # 실제 요청된 URL 확인
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">HTML 파싱 (BeautifulSoup)</span>

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(res.text, "html.parser")
print(soup.title.text)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">태그로 데이터 추출</span>

```python
links = soup.find_all("a")

for link in links[:5]:
    print(link.text)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">CSS 선택자로 추출</span>

```python
items = soup.select("a")

for i in items[:5]:
    print(i.get_text())
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">전체 흐름(미니 크롤링)</span>

```python
import requests
from bs4 import BeautifulSoup

url = "https://example.com"

res = requests.get(url)
soup = BeautifulSoup(res.text, "html.parser")

titles = soup.find_all("h1")

for t in titles:
    print(t.text)
```

📌 한눈에 보는 핵심 구조  
requests → 페이지 요청  
status_code → 성공 확인  
headers → 차단 방지  
params → 검색/조건  
BeautifulSoup → 파싱  
find / select → 데이터 추출
