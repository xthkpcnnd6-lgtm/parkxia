<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">배열 생성과 난수</span>

```python
import numpy as np

# 1~100 사이의 3의 배수 배열 생성
three_multiples = np.arange(3, 101, 3)
print("3의 배수:", three_multiples)

# 평균 0, 표준편차 1인 표준정규분포 난수 5개 생성[cite: 1]
normal_dist = np.random.randn(5)
print("표준정규분포 난수:", normal_dist)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">불리언 인덱싱</span>

```python
# 0~10 사이의 정수 10개 생성[cite: 1]
arr = np.random.randint(0, 11, 10)

# 5보다 큰 요소만 추출[cite: 1]
over_five = arr[arr > 5]

# 3보다 크고 8보다 작은 요소 추출 (복합 조건)[cite: 1]
filtered = arr[(arr > 3) & (arr < 8)]

print(f"원본: {arr}\n5초과: {over_five}\n3~8사이: {filtered}")
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">차원 변경과 슬라이싱</span>

```python
# 1차원 배열을 2x5 행렬로 변경[cite: 1]
matrix = np.arange(10).reshape(2, 5)

# 전체 행에서 0~2열까지만 슬라이싱[cite: 1]
sub_matrix = matrix[:, :3]

print("2x5 행렬:\n", matrix)
print("슬라이싱 결과:\n", sub_matrix)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">브로드캐스팅</span>

```python
# (3, 3) 행렬 생성[cite: 1]
grid = np.zeros((3, 3))

# (3,) 크기의 1차원 배열(벡터) 더하기[cite: 1]
# 각 행마다 [1, 2, 3]이 더해집니다.
row_add = grid + np.array([1, 2, 3])

print("브로드캐스팅 결과:\n", row_add)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">view vs copy 주의사항</span>

```python
original = np.array([10, 20, 30, 40])

# 슬라이싱 (View 생성)[cite: 1]
view_copy = original[:2]
view_copy[0] = 99

print("원본(수정됨):", original) # 원본의 10이 99로 바뀝니다![cite: 1]

# .copy() 사용 (독립적 복사본)[cite: 1]
real_copy = original[:2].copy()
real_copy[1] = 777

print("원본(유지됨):", original) # 원본의 20은 그대로 유지됩니다.[cite: 1]
```
