<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">퍼셉트론(Perceptron)</span>

```python
import numpy as np

# 입력
x = np.array([1, 2])
w = np.array([0.5, 0.5])
b = -1

# 퍼셉트론
y = np.dot(x, w) + b
output = 1 if y > 0 else 0

print(output)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">퍼셉트론 분류 경계</span>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 5, 100)
y = -x + 2  # 경계선

plt.plot(x, y)
plt.title("Perceptron Decision Boundary")
plt.show()
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">활성화 함수</span>

```python
import numpy as np

def step(x):
    return np.where(x > 0, 1, 0)

print(step(np.array([-1, 0, 1, 2])))
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">다층 퍼셉트론(MLP)</span>

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

model = MLPClassifier(hidden_layer_sizes=(5,))
model.fit(X, y)

print(model.score(X, y))
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">순전파(Forward Propagation)</span>

```python
import numpy as np

x = np.array([1, 2])
w = np.array([[0.5, 0.2], [0.3, 0.8]])
b = np.array([0.1, 0.1])

z = np.dot(x, w) + b
print(z)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">손실 함수 (Loss)</span>

```python
import numpy as np

y_true = np.array([1])
y_pred = np.array([0.8])

loss = np.mean((y_true - y_pred) ** 2)
print(loss)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">역전파 개념 (Gradient 느낌)</span>

```python
import numpy as np

w = 0.5
x = 2
y_true = 1

y_pred = w * x
loss = (y_true - y_pred) ** 2

# 간단한 기울기
grad = -2 * x * (y_true - y_pred)

print("gradient:", grad)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">가중치 업데이트 (학습)</span>

```python
learning_rate = 0.1

w = w - learning_rate * grad
print("updated weight:", w)
```
