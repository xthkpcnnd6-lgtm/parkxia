<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">퍼셉트론(기본 구조)</span>

```python
import numpy as np

x = np.array([1, 2])
w = np.array([0.5, 0.5])
b = -1

y = np.dot(x, w) + b
print(1 if y > 0 else 0)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">활성화 함수(Sigmoid)</span>

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

print(sigmoid(np.array([-1, 0, 1])))
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">ReLU 함수</span>

```python
import numpy as np

def relu(x):
    return np.maximum(0, x)

print(relu(np.array([-1, 0, 2])))
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">다층 퍼셉트론(MLP)</span>

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import load_iris

X, y = load_iris(return_X_y=True)

model = MLPClassifier(hidden_layer_sizes=(10,))
model.fit(X, y)

print(model.score(X, y))
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">순전파(Forward)</span>

```python
import numpy as np

x = np.array([1, 2])
w = np.array([[0.5, 0.2], [0.3, 0.8]])
b = np.array([0.1, 0.1])

z = np.dot(x, w) + b
print(z)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">손실 함수(MSE)</span>

```python
import numpy as np

y_true = np.array([1])
y_pred = np.array([0.8])

loss = np.mean((y_true - y_pred) ** 2)
print(loss)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">BCE 손실 함수(이진 분류)</span>

```python
import numpy as np

y = 1
p = 0.8

loss = -(y*np.log(p) + (1-y)*np.log(1-p))
print(loss)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">간단한 역전파(Gradient)</span>

```python
w = 0.5
x = 2
y_true = 1

y_pred = w * x
grad = -2 * x * (y_true - y_pred)

print(grad)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">가중치 업데이트</span>

```python
lr = 0.1
w = w - lr * grad

print(w)
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">이진 분류(Logistic Regression)</span>

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X, y)

print(model.predict(X[:5]))
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">케라스 모델 만들기</span>

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy')
print("모델 생성 완료")
```

<span style="font-family: 'Gungsuh', '바탕체'; font-size: 2em; font-style: italic;">파이토치 기본 모델</span>

```python
import torch.nn as nn

model = nn.Sequential(
    nn.Linear(2, 10),
    nn.ReLU(),
    nn.Linear(10, 1)
)

print(model)
```
